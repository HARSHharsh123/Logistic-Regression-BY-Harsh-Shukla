## Fully Updated Perceptron Trick having Activation Function as Sigmoid and Loss function as Log loss function

```python
def GradientDescent(X, y):
    X = np.insert(X , 0 , 1, axis = 1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    for i in range(5000):
        y_hat = sigmoid(np.dot(X , weights))
        weights = weights + lr * (np.dot((y - y_hat) , X)) / X.shape[0]
    return weights[1:] , weights[0]
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
coef_1 , intercept_1= GradientDescent(X,y)
```

```python
coef_1 , intercept_1
```

```python
m1 = -(coef_1[0]  / coef_1[1])
c1 = -(intercept_1 / coef_1[1])
```

```python
x_input1 = np.linspace(-3,2,100)
y_input1 = m1 * x_input1 + c1
```

```python
plt.plot(x_input1 , y_input1, color = 'r' , linewidth = 3)
plt.scatter(x = X[:,0] , y = X[:,1] , c = y , cmap = 'winter') ## from matplotlib
plt.ylim(-3,2)
```
