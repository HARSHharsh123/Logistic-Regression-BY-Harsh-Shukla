# Updation in perceptron trick to build Logistic Regression 

```python
from sklearn.datasets import make_classification
X , y = make_classification(n_samples = 100 , n_features = 2 , n_classes = 2 , n_clusters_per_class=1 , n_informative = 1 , n_redundant = 0 ,
                           random_state = 41 , class_sep = 20 , hypercube= False)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.scatter(x = X[:,0] , y = X[:,1] , c = y , cmap = 'winter') ## from matplotlib
```

## Perceptron Trick with Activation function as Step Function

```python
## perceptron trick with Step Function (previous Concept)

import numpy as np
def Perceptron(X , y):
    X = np.insert(X , 0 , 1 , axis = 1)
    w = np.ones(X.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j] , w))
        w = w + lr * (y[j] - y_hat) * X[j]
    return w[0] , w[1:]
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
intercept_, coef_ = Perceptron(X,y)
```

```python
intercept_
```

```python
coef_
```

```python
m = -(coef_[0] / coef_[1])
c = -(intercept_ / coef_[1])
```

```python
m , c
```

```python
x_input = np.linspace(-3,2,100)
y_input = m*x_input + c
```

```python
plt.plot(x_input , y_input, color = 'r' , linewidth = 3)
plt.scatter(x = X[:,0] , y = X[:,1] , c = y , cmap = 'winter') ## from matplotlib
plt.ylim(-3,2)
```

## Updated Perceptron Trick with Activation Function as Sigmoid Function

```python
## perceptron trick with Sigmoid Function 

import numpy as np
def Perceptron(X , y):
    X = np.insert(X , 0 , 1 , axis = 1)
    w = np.ones(X.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = sigma(np.dot(X[j] , w))
        w = w + lr * (y[j] - y_hat) * X[j]
    return w[0] , w[1:]
```

```python
def sigma(z):
    return 1 / (1 + np.exp(-z))
```

```python
intercept_1 , coef_1 = Perceptron(X,y)
```

```python
intercept_1
```

```python
coef_1
```

```python
m1 = -(coef_1[0] / coef_1[1])
c1 = -(intercept_1 / coef_1[1])
```

```python
x_input1 = np.linspace(-3,2,100)
y_input1 = m1*x_input1 + c1
```

```python
plt.plot(x_input1 , y_input1, color = 'brown' , linewidth = 3)
plt.scatter(x = X[:,0] , y = X[:,1] , c = y , cmap = 'winter') ## from matplotlib
plt.ylim(-3,2)
```

## Logistic Regression Algorithm by Sklearn

```python
intercept_2 = clf.intercept_
```

```python
coef_2 = clf.coef_
```

```python
m2 = -(coef_2[0][0] / coef_2[0][1])
c2 = -(intercept_2 / coef_2[0][1])
```

```python
x_input2 = np.linspace(-3,2,100)
y_input2 = m2*x_input2 + c2
```

```python
plt.plot(x_input2 , y_input2, color = 'black' , linewidth = 3)
plt.scatter(x = X[:,0] , y = X[:,1] , c = y , cmap = 'winter') ## from matplotlib
plt.ylim(-3,2)
```
